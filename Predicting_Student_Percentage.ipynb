# Predicting Student Percentage based on the number of Study Hours.

This is very simple Machine learning project based on Linear Regression Model under supervised machine learning algorithms category.


```python
#  Libraries

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Fetching data from .csv file into pandas dataframe.
df = pd.read_csv('studentData_HoursAndScores.csv')
print(df)
```

# some analysis of data

```python
# data type of each attribute in a DataFrame
print(df.dtypes)
```

```python
df.describe()
```

# Visualisation of data

```python
# Visualisation using scatter plot.

plt.title('Student study hours with score ')
plt.scatter(df['Hours'],df['Scores'])
plt.xlabel('Study Hours')
plt.ylabel('Score')
plt.show()
```

We can see that with the increase in study hours, the score also increases.

From the graph above, we can clearly see that there is a positive linear relation between the number of hours studied and percentage of score.

so here Simple Linear regression model will be suitable for this type of dataset.

y = A(x) + B

y is dependent variable i.e. Score

x is independent variable i.e. Study Hours

# Preparing the Data

```python
# X = df.iloc[:,:-1].values
# print(X)

# Y = df.iloc[:,1:2].values
# print(Y)
```

```python
# Study Hours
X1 = df['Hours'].values
X1 = X1.reshape(-1,1) # need to create 2-D Array instead of 1-D Array
print(X1)
```

```python
# Score
Y1 = df['Scores'].values
Y1 = Y1.reshape(-1,1) # need to create 2-D Array instead of 1-D Array
print(Y1)
```

Split this data into training and test sets. We'll do this by using Scikit-Learn's built-in train_test_split() method:

```python
X_train, X_test, Y_train, Y_test = train_test_split(X1,Y1,test_size=0.2,train_size=0.8,random_state=0)
```

Here we will be using 80% of data from dataset from training the model using train_size=0.8 argument of the train_test_split().

Here we will be using 20% of data from dataset from testing the model using test_size=0.2 argument of the train_test_split().

# Training the Model.

```python
# Linear Regression model.

reg = LinearRegression()
reg.fit(X_train,Y_train)
```

```python
line = (reg.coef_ * X1) + reg.intercept_ # is also equivalent to reg.predict(X)

# Visualisation using scatter plot.

plt.title('Student study hours with score ')
plt.scatter(X1,Y1) 
# plt.scatter(df['Hours'],df['Scores'])
plt.plot(X1,line)
plt.xlabel('Study Hours')
plt.ylabel('Score')
plt.show()
```

# Testing

```python
print(X_test)
```

```python
Y_pred = reg.predict(X_test)
print(Y_pred)
```

Comparing actual result and Predicted result using linear model.

```python
print('Actual score: -')
print(Y_test)
print()
print('Predict score: -')
print(Y_pred)
```

# Predicting the values

What will be predicted score if a student studies for 9.25 hrs/day ?

```python
hrs = input("Enter the Students Study Hours: - ")
hrs = float(hrs)
score = reg.predict([[hrs]])
print('Expected Score is %f' % score)
```
